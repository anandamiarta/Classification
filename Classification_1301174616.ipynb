Mengklasifikasi label data overall players yang udah di kategorikan menjadi 3 label berdasarkan data players menggunakan Algoritma KNN dan Naive Bayes


IMPORT LIBRARY

```
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
```

MEMANGGIL DATA FIFA20.CSV

```
df = pd.read_csv('fifa20.csv') #read data
df.head() #munculkan data
```

DROP FITUR DATA YANG TIDAK DIPERLUKAN

```
#menghapus variable data yang tidak digunakan
df = df.drop(columns=['ls',	'st',	'rs',	'lw',	'lf',	'cf',	'rf'	,'rw',	'lam',	'cam',	'ram',	'lm',	'lcm',	'cm',	'rcm',	'rm',	'lwb',	'ldm',	'cdm'	,'rdm',	'rwb',	'lb',	'lcb',	'cb',	'rcb',	'rb',
	'player_positions',	'international_reputation',	'weak_foot', 'team_position',	'team_jersey_number',	'contract_valid_until',	'nation_position',	
'nation_jersey_number',	'wage_eur','work_rate',	'release_clause_eur',	
'joined', 'player_traits', 'player_tags','real_face', 'preferred_foot', 'loaned_from', 'long_name','nationality', 'club','value_eur','short_name',
'age',	'height_cm',	'weight_kg',	'body_type','dob','player_url','sofifa_id','gk_kicking'])
df.head()
```

MENGISI DATA YANG KOSONG

```
nums = ['physic','pace'	,'shooting',	'passing',	'dribbling',	'defending',	'gk_diving',	'gk_handling',	'gk_reflexes',	'gk_speed',	'gk_positioning']
cats = df.columns.drop(nums)
for numer in nums:
     df[numer] = df[numer].fillna(df[numer].mean()) #isi data NAN
for cate in cats:
     df[cate] = df[cate].fillna(df[cate].mode().values[0]) #isi data NAN
df.head()
```

```
df['overall'].value_counts()
```

MENGKATEGORIKAN DATA YANG DIBAGI MENJADI 3 LABEL AGAR LEBIH AKURAT

```
df.loc[df['overall'] > 80, 'overall'] = 90 #Data dikategorikan ke 90 jika nilai data > 80
df.loc[((df['overall'] <= 80) & (df['overall'] > 60)), 'overall'] = 70 #Data dikategorikan ke 70 jika nilai data <= 80 dan nilai data > 60
df.loc[df['overall'] <= 60, 'overall'] = 50 #Data dikategorikan ke 60 jika nilai data > 60

df.loc[df['overall'] == 90 , 'overall'] = 2 #Data dikategorikan ke 2 jika nilai data = 90
df.loc[df['overall'] == 70 , 'overall'] = 1 #Data dikategorikan ke 1 jika nilai data = 70
df.loc[df['overall'] == 50 , 'overall'] = 0 #Data dikategorikan ke 0 jika nilai data = 50

df
```

HITUNG JUMLAH LABEL DATA YANG FIX DARI OVERALL

```
df['overall'].value_counts() #hitung jumlah label data
```

MENENTUKAN BODY_TYPE SEBAGAI LABEL

```
#overall sebagai label
features = df.drop(columns=['overall']).values 
target =  df.overall

#Train Test Split untuk membagi dataset menjadi training set dan test set
features_train, features_test, target_train, target_test = train_test_split(features,target, test_size = 0.3, train_size = 0.7, random_state = 100)
```

EXPERIMENT 1 MENGGUNAKAN ALGORITMA NAIVE BAYES DENGAN DATA TEST

```
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
gnb = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
y_pred = gnb.fit(features_train, target_train).predict(features_test)

from sklearn.metrics import accuracy_score
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(target_test,y_pred))
print("Number of mislabeled points out of a total %d points : %d"% (features_test.shape[0], (target_test != y_pred).sum()))
accuracy = accuracy_score(target_test, y_pred, normalize = True)
print(accuracy)
```

EXPERIMENT 1.1 MENGGUNAKAN ALGORITMA NAIVE BAYES DENGAN DATA TRAIN

```
# Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
# Mengaktifkan/memanggil/membuat fungsi klasifikasi Naive bayes
gnb = GaussianNB()
# Memasukkan data training pada fungsi klasifikasi naive bayes
y_pred = gnb.fit(features_train, target_train).predict(features_train)

from sklearn.metrics import accuracy_score
# Menghitung nilai akurasi dari klasifikasi naive bayes 
from sklearn.metrics import classification_report
print(classification_report(target_train,y_pred))
print("Number of mislabeled points out of a total %d points : %d"% (features_train.shape[0], (target_train != y_pred).sum()))
accuracy = accuracy_score(target_train, y_pred, normalize = True)
print(accuracy)
```

EXPERIMENT 2 MENGGUNAKAN ALGORITMA KNN DENGAN DATA SET

```
from sklearn.neighbors import KNeighborsClassifier
# Mengaktifkan fungsi klasifikasi
klasifikasi = KNeighborsClassifier(n_neighbors=100)
# Memasukkan data training pada fungsi klasifikasi
klasifikasi.fit(features_train, target_train)
# Menentukan hasil prediksi dari features_test
y_pred = klasifikasi.predict(features_test)
y_pred

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
# Menghitung nilai akurasi dari klasifikasi KNN
print(classification_report(target_test, y_pred))
print("Number of mislabeled points out of a total %d points : %d"% (features_test.shape[0], (target_test != y_pred).sum()))
accuracy= accuracy_score(target_test, y_pred)
accuracy
```
